```python
import numpy as np
import pandas as pd
import cv2
import glob
import os
```

```python
pth_= 'E:/KHA240214/segmentstion_task/Reham/images/'
for img in glob.glob(f'{pth_}*.png'):
    print(img)
    _, im = os.path.split(img)
    imname = im.split('.png')[0]
    image = cv2.imread(img)
    w, h, _ = image.shape
    print(w, h)
    for x in range(0,w,256):
        for y in range(0,h,256):
            crp_im = image[y:y+512, x:x+512, :]
            new_name = pth_ + 'crp/' + imname + f'_{x}_{y}.png'
            #print(crp_im.size)
            if crp_im.size != 0:
                cv2.imwrite(new_name, crp_im)
```

```python
pth_= 'E:/KHA240214/Reham/PT+PBM 28_CD31/CD31/'
for img in glob.glob(f'{pth_}*.png'):
    print(img)
    _, im = os.path.split(img)
    imname = im.split('.png')[0]
    image = cv2.imread(img)
    w, h, _ = image.shape
    print(w, h)
    patch = 512

    newpth_= 'E:/KHA240214/Reham/PT+PBM 28_CD31/test2/'

    for x in range(0,w,patch):
        if x+patch> w:
            x = w-patch
        for y in range(0,h,patch):
            if y+patch> h:
                y = h-patch
            crp_im = image[x:x+patch, y:y+patch, :]
            new_folder = newpth_ + imname 
            new_name = new_folder +'/'+ imname + f'_{x}_{y}.jpg'
            if not os.path.exists(new_folder):
                os.makedirs(new_folder)
            #print(crp_im.size)
            if crp_im.size != 0:
                cv2.imwrite(new_name, crp_im)
```
